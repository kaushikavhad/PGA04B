```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
#.docx file saved as .txt and reading  txt file
f = open("M:\\Imarticus\\data sets\\Email archive for text mining.txt", "r")
x=f.read()
f.close() 
```

```python
#Removing stopwords
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(x)
filtered_sentence = [w for w in word_tokens if not w in stop_words]
```

```python
#stemmind words
from nltk.stem import PorterStemmer
ps = PorterStemmer()   
stemmed_words=[]
for w in filtered_sentence:
    stemmed_words.append(ps.stem(w))
```

```python
#Lemmatizing words   
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()
lem_word=[]
for w in stemmed_words:
    lem_word.append(lem.lemmatize(w,"v"))  
```

```python
str = " "
string_convertion=str.join(lem_word)

#remove special char. and numbers    
import re
cleanString = re.sub('\W+',' ',string_convertion )
string_no_numbers = re.sub("\d+", " ", cleanString)
word_tokens = word_tokenize(string_no_numbers)
```

```python
#remove own stopwords
stop_words_lst = ['also', 'share','P','G','To','Cc','lc','arx','soon','mayb',
                  'lcp','I','would','from','oom','s']
rm_stopwords= [w for w in word_tokens if not w in stop_words_lst]
```

```python
#word freq plot
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style('darkgrid')
nlp_words=nltk.FreqDist(rm_stopwords)
nlp_words.plot(50)
```

```python
#I have collected words related to decision in file decision_synonyms.txt 

syn_file= open("M:\Imarticus\data sets\Text mining\decision_synonyms.txt", "r")
synonyms=syn_file.read()
syn_file.close() 
```

```python
# If words in file decision_synonyms.txt == words in file Email archive for text mining.txt
# collecting that words in variable decision_syn

syn_string = re.sub("\n", " ", synonyms)
word_tokens4= word_tokenize(syn_string)
decision_syn= [w for w in rm_stopwords if w in word_tokens4]
```

```python
# word freq plot for words related to decision in file.
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style('darkgrid')
nlp_words=nltk.FreqDist(decision_syn)
nlp_words.plot(100)
```
