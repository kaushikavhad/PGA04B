```r
credit_data_sample<- read.csv("M:\\Imarticus\\ProjectR\\Credit card fraud\\Datafiles\\credit_data_sample.csv", header = TRUE,
                              stringsAsFactors = TRUE,
                              strip.white = FALSE ,
                              sep = ',')
```

```r
credit_data_sample$Class= as.factor(credit_data_sample$Class)

library(party)
library(partykit)

credit_ctree <- ctree(Class ~., data=credit_data_sample)
credit_ctree1 <- ctree(Class ~V14+V4, data=credit_data_sample)
credit_ctree

```

```r
plot(as.simpleparty(credit_ctree), type="simple")
```

```r
plot(as.simpleparty(credit_ctree1), type="simple")
```

```r
library(tree)
credit_tree <- tree(Class ~., data=credit_data_sample)
plot(credit_tree)
text(credit_tree)
```

```r
library(rpart)
library(rattle)

rpart <- rpart(Class ~., data=credit_data_sample, method="class")
fancyRpartPlot(rpart)
```
