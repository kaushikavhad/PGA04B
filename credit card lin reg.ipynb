```r
library(readr)
credit_data<- read.csv("M:\\Imarticus\\data sets\\Text mining\\creditcard.csv", header = TRUE,
                       stringsAsFactors = TRUE,
                       strip.white = FALSE ,
                       sep = ',')
```

```r
library(dplyr)
credit_data_Lreg<- select(credit_data,c(-31))
head(credit_data_Lreg)
```

```r
cor(credit_data_Lreg)
```

```r
summary(credit_data_Lreg)
```

```r
linearmod<- lm(Amount~.,data=credit_data_Lreg)
summary(linearmod)
```

```r
linearmod<- lm(Amount~V1+V2+V3+V5+V6+V7+V20,data=credit_data_Lreg)

summary(linearmod)
```

```r
library(caTools)

split = sample.split(credit_data_Lreg$Amount, SplitRatio = .8)
training_data = subset(credit_data_Lreg, split == TRUE)
test_data = subset(credit_data_Lreg, split == FALSE)
str(test_data)
```

```r
lin_mod_train<- lm(Amount~V1+V2+V3+V5+V6+V7+V20,data=training_data)
summary(lin_mod_train)
```

```r
pred_dist=predict(lin_mod_train,test_data)
result<- data.frame(Amount=test_data$Amount,pred_Amount=pred_dist,diff=(test_data$Amount - pred_dist))
result$ per.error<- (result$diff/result$Amount)*100
head(result)
```
